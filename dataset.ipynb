```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python

x = np.random.normal(size=10000)


sns.set_style('ticks')


sns.displot(x, kde=True)

# Mean and standard deviation
print("Mean:", x.mean())
print("Std:", x.std())
```

```python
#RANDOM SAMPLE

x_sample = np.random.choice(x, size=10, replace=False)

print("Sample:", x_sample)

print("Sample Mean:", x_sample.mean())
```

```python
# Function to calculate sample means
def sample_mean_calculator(input_dist, sample_size, n_samples):
    sample_means = []

    for i in range(n_samples):
        sample = np.random.choice(
            input_dist,
            size=sample_size,
            replace=False
        )

        sample_means.append(sample.mean())

    return sample_means
```

```python
#Small Number of Samples
sns.displot(
    sample_mean_calculator(x, 10, 10),
    color='green',
    kde=True
)

plt.xlim(-1.5, 1.5)
plt.show()
```

```python
#More Samples
sns.displot(
    sample_mean_calculator(x, 10, 1000),
    color='green',
    kde=True
)

plt.xlim(-1.5, 1.5)
plt.show()
```

```python
#Larger Sample Size
sns.displot(
    sample_mean_calculator(x, 100, 1000),
    color='green',
    kde=True
)

plt.xlim(-1.5, 1.5)
plt.show()
```

```python
#Even Larger Sample Size
sns.displot(
    sample_mean_calculator(x, 1000, 1000),
    color='green',
    kde=True
)

plt.xlim(-1.5, 1.5)
plt.show()
```

```python
# Skewed distribution
s = st.skewnorm.rvs(10, size=10000)


sns.displot(s, kde=True)

```

```python
#Sample Means from Skewed Data
sns.displot(
    sample_mean_calculator(s, 1000, 1000),
    color='green',
    kde=True
)

plt.show()
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Sample Dataset
data = np.array([85, 90, 78, 92, 88, 76, 95, 89, 82, 80])

# Mean and Standard Deviation
mean_val = np.mean(data)
std_val = np.std(data)

print(f"Mean: {mean_val:.2f}")
print(f"Standard Deviation: {std_val:.2f}")

# Z-scores for all data points
z_scores = stats.zscore(data)
print("Z-scores:")
print(z_scores)

# Finding X from Z = 1.2
z_input1 = 1.2
x_val1 = mean_val + (z_input1 * std_val)

print(f"\nRaw score (X) for Z = 1.2: {x_val1:.2f}")

# Finding X from Z = -1.5
z_input2 = -1.5
x_val2 = mean_val + (z_input2 * std_val)

print(f"Raw score (X) for Z = -1.5: {x_val2:.2f}")

# Histogram
plt.hist(data, bins=5)
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Histogram of Dataset")
plt.show()
```
